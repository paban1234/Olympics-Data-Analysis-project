```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Load dataset
df = pd.read_csv("Summer-Olympic-medals-1976-to-2008.csv", encoding='latin1')
df.head()
```

```python
# Drop columns we don't need
df.drop(['Event_gender', 'Country_Code'], axis=1, inplace=True)

# Drop rows where all values are missing
df.dropna(how='all', inplace=True)

# Convert 'Year' to integer
df['Year'] = df['Year'].astype(int)

# Check the cleaned data
df.info()
df.head()
```

```python
# Count number of events hosted by each city
city_counts = df['City'].value_counts()

# Plot
plt.figure(figsize=(12, 5))
city_counts.plot(kind='bar', color='skyblue')
plt.title("Number of Olympic Events Hosted by City")
plt.xlabel("City")
plt.ylabel("Number of Events")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count total medals per country
country_medal_count = df['Country'].value_counts()

# Plot
plt.figure(figsize=(10, 5))
country_medal_count.head(10).plot(kind='bar', color='gold')
plt.title("Top 10 Countries by Total Medals")
plt.xlabel("Country")
plt.ylabel("Medal Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by Year and count medals
medals_by_year = df.groupby('Year')['Medal'].count()

# Plot
plt.figure(figsize=(10, 5))
plt.plot(medals_by_year.index, medals_by_year.values, marker='o', linestyle='-', color='green')
plt.title("Total Medals Awarded Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Medals")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Count male vs female
gender_counts = df['Gender'].value_counts()

# Pie chart
plt.figure(figsize=(6, 6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['#66b3ff', '#ff9999'], startangle=140, explode=[0.05, 0])
plt.title("Gender Distribution in Medal Winners")
plt.ylabel("")
plt.show()
```

```python
# Count medals per athlete
top_athletes = df['Athlete'].value_counts().head(10)

# Plot
plt.figure(figsize=(12, 6))
top_athletes.plot(kind='bar', color='silver')
plt.title("Top 10 Athletes by Total Medals")
plt.xlabel("Athlete")
plt.ylabel("Number of Medals")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
